```python
%%sql

DROP TABLE IF EXISTS POS_TRX;
CREATE TABLE POS_TRX(ID_TRX INT PRIMARY KEY, Tienda VARCHAR(255), Fecha_transaccion DATE, Tipo_transaccion INT)
```

```python
%%sql

DROP TABLE IF EXISTS POS_TRX_PAGO;
CREATE TABLE POS_TRX_PAGO(ID_TRX INT PRIMARY KEY, Monto_pago INT, Medio_de_pago VARCHAR(45))
```

```python
%%sql

DROP TABLE IF EXISTS POS_TRX_CLIENTE;
CREATE TABLE POS_TRX_CLIENTE(ID_TRX INT PRIMARY KEY, Nombre VARCHAR(45), Rut VARCHAR(45))
```

```python
# Escriba una consulta en SQL que retorne la cantidad de clientes distintos por tienda que realizaron una transacción del tipo 15.

%%sql 

SELECT POS_TRX.Tienda, COUNT(DISTINCT POS_TRX_CLIENTE.Rut) Clientes
FROM POS_TRX, POS_TRX_CLIENTE
WHERE POS_TRX.ID_TRX = POS_TRX_CLIENTE.ID_TRX AND POS_TRX.Tipo_transaccion = 15
GROUP BY POS_TRX.Tienda
```

```python
# Escriba una consulta en SQL que entregue las ventas de cada tienda por año en orden de mayor a menor.

%%sql 

SELECT POS_TRX.Tienda, strftime('%Y', POS_TRX.Fecha_transaccion) Año, SUM(POS_TRX_PAGO.Monto_pago) Monto
FROM POS_TRX, POS_TRX_PAGO
WHERE POS_TRX.ID_TRX = POS_TRX_PAGO.ID_TRX
GROUP BY POS_TRX.Tienda, Año
HAVING SUM(POS_TRX_PAGO.Monto_pago)
ORDER BY Tienda Asc, Año Desc
```

```python
#Escriba una consulta en SQL que entregue el valor de la transacción promedio de cada tienda por año en orden de mayor a menor.

%%sql 

SELECT POS_TRX.Tienda, strftime('%Y', POS_TRX.Fecha_transaccion) Año, AVG(POS_TRX_PAGO.Monto_pago) Monto_medio
FROM POS_TRX, POS_TRX_PAGO
WHERE POS_TRX.ID_TRX = POS_TRX_PAGO.ID_TRX
GROUP BY POS_TRX.Tienda, Año
ORDER BY Tienda Asc, Año Desc
```

```python
# Escriba una consulta en SQL que responda: ¿Cuál es el principal medio de pago? 

#Según número de transacciones

%%sql 

SELECT POS_TRX_PAGO.Medio_de_pago Medio, COUNT(POS_TRX_PAGO.Medio_de_pago) Transacciones
FROM POS_TRX_PAGO
GROUP BY POS_TRX_PAGO.Medio_de_pago
ORDER BY COUNT(POS_TRX_PAGO.Medio_de_pago) Desc
Limit 1


```
